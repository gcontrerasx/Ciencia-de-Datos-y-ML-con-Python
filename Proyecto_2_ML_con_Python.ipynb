```python
#importacion de libreria PANDAS
import pandas as pd
```

```python
#Cargamos los dados del DATASET, aca indicamos que el separador es punto y coma ";"
data_autos = pd.read_csv('autos.csv',sep=";")
data_autos.head()
```

```python
#Verificamos los datos, cuantos datos nulos hay en el DATA SET
data_autos.info()
```

```python
#Verificamos cantidad de datos nulos en el DATA SET el % es bajo, 14 datos nulos:
data_autos.isnull().sum()
```

```python
#Etiquetamos la variable: Pais_Origen: USA, Europe, Japan a la nueva variabla, Origen_tipo: 1, 2, 3:  
data_autos["Origen_Tipo"]=data_autos["Pais_Origen"].apply(lambda x: "1" if x == "USA" else ("2" if x == "Europe" else ("3" if x == "Japan" else "" )))
data_autos
```
